```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
```

# HTML 문서 불러오기

```python
driver = webdriver.Chrome()
# time.sleep(3)

driver.get('https://news.google.com/covid19/map?hl=ko&gl=KR&ceid=KR%3Ako')
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

covid_list = soup.select('tr.sgXwHf.wdLSAe')
print(covid_list[0].prettify())

driver.close()
```

```python
covid = covid_list[0]
for c in covid:
    print(c.text)
```

```python
# parsing text 
covid_data = [c.text for c in covid_list[0].select('.l3HOY')]

# unpack covid
location, total, new, _, per_1m, deaths = covid_data

print(f'Location: {location}')
print(f'Total cases: {total}')
print(f'New cases: {new}')
print(f'Cases per 1m: {per_1m}')
print(f'Deaths: {deaths}')
```

```python
data = []

for covid_info in covid_list:
    covid_data = [c.text for c in covid_info.select('.l3HOY')]
    location, total, new, _, per_1m, deaths = covid_data
    
    data.append([location, total, new, per_1m, deaths])
    
#     print('-'*50)
#     print(f'Location: {location}')
#     print(f'Total cases: {total}')
#     print(f'New cases: {new}')
#     print(f'Cases per 1m: {per_1m}')
#     print(f'Deaths: {deaths}')

pd.DataFrame(data, columns=['location', 'total', 'new', 'per_1m', 'deaths'])
# 지금 페이지에서 idx는 쓰이지 않는 것 같은데 enumerate 함수를 쓰신 이유가 있으신가요???
```
