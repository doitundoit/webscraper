```python
from selenium import webdriver
from bs4 import BeautifulSoup as bs
from IPython.display import display
import pandas as pd
import time
```

## 댓글

```python
driver = webdriver.Chrome()
driver.get('https://news.v.daum.net/v/20210815063015157')
time.sleep(3)

html = driver.page_source
soup = bs(html, 'html.parser')

driver.close()
```

```python
cmt_count = soup.select('#alex-header > em')
cmt_box = soup.select('#alex-area > div > div > div > div.cmt_box > ul.list_comment > li > div')
cmt_box

print(f'댓글: {cmt_count[0].text}')

cmts = []
for cmt in cmt_box:
    author = cmt.select('strong > span > a > span:nth-child(2)')[0].text
    txt = cmt.select('p')[0].text.replace('\n', ' ')
    cmts.append([author, txt])
    
df = pd.DataFrame(cmts, columns=['작성자', '댓글'])
df
```

## 댓글 더보기

```python
# options = webdriver.ChromeOptions()
# options.headless = True

# driver = webdriver.Chrome(options = options)
driver = webdriver.Chrome()
driver.get('https://news.v.daum.net/v/20210815063015157')
time.sleep(3)

more_btn = driver.find_element_by_xpath('//*[@id="alex-area"]/div/div/div/div[3]/div[3]/button')
more_btn.click()
time.sleep(3)

html = driver.page_source
soup = bs(html, 'html.parser')
driver.implicitly_wait(3)
driver.close()
```
