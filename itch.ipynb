## itch 스크래핑

```python
# headless 옵션
# options = webdriver.ChromeOptions()
# options.headless = True

# driver 초기화
# driver = webdriver.Chrome(options=options)
driver = webdriver.Chrome()
driver.get('https://itch.io/games/new-and-popular')

# html 파싱
html = driver.page_source
soup = bs(html, 'html.parser')
soup
```

```python
data = []
game_cells = soup.select('div.game_cell_data')

for game_cell in game_cells:
    
    title = game_cell.select('.game_title')
    author = game_cell.select('.game_author')
    genre = game_cell.select('.game_genre')
    price = game_cell.select('.price_value')
    platform = game_cell.select('.game_platform > span')
    
    if len(genre) != 0:
        genre = genre[0].text
        
    else:
        genre = 'null'
    
    if len(price) != 0:
        price = price[0].text.lstrip('$')
        
    else: 
        price = 'null'
    
    data.append( [title[0].text, author[0].text, genre, price] )
    
itch = pd.DataFrame(data, columns=['title', 'author', 'genre', 'price'])
itch
```
