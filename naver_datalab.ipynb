# 네이버 데이터랩 스크래핑

```python
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import pandas as pd
```

## webdriver 생성 및 초기화

```python
# headless option
options = webdriver.ChromeOptions()
options.headless = True

# driver 오브젝트 생성
driver = webdriver.Chrome(options = options)

# 페이지 로딩 3초
driver.implicitly_wait(3)
```

## html 호출 및 파싱

```python
# html request
driver.get('https://datalab.naver.com/home/sectionSearch.naver')

# driver의 html 문서를 받아서 soup로 파싱
html = driver.page_source
soup = bs(html, 'html.parser')
```

## 텍스트 추출

```python
selector = '#content > div.spot.section_keyword > div.home_section.active > div > div.keyword_carousel > div > div > div:nth-child(12) > div > div > ul > li > a > span'
ranks = soup.select(selector)
ranks_keyword = [ rank.text for rank in ranks ]
ranks_num = [num for num in range(1, 11)]
```

## pandas 테이블 생성

```python
df = pd.DataFrame(data={'rank':ranks_num, 'keyword':ranks_keyword})
df = df.set_index('rank')
print(df)
```

## 종료

```python
driver.quit()
```
